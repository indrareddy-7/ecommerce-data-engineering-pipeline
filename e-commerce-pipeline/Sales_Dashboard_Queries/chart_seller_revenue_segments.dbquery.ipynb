```sql
SELECT
CASE
WHEN seller_revenue >= 200000 THEN 'High Performing Sellers'
WHEN seller_revenue BETWEEN 100000 AND 199999 THEN 'Medium Performing Sellers'
ELSE 'Low Performing Sellers'
END AS seller_segment,
COUNT(*) AS seller_count
FROM (
SELECT
seller_id,
SUM(price + freight_value) AS seller_revenue
FROM main_workspace.gold.fact_sales
GROUP BY seller_id
) s
GROUP BY seller_segment
ORDER BY seller_segment DESC
```
