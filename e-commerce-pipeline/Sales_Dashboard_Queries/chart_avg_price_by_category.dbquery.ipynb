```sql
SELECT
CONCAT(
    INITCAP(REPLACE(p.product_category_name,'_',' ')),
    '  ( ',
    INITCAP(REPLACE(t.product_category_name_english,'_',' ')),
    ' )'
) AS product_category,
ROUND(AVG(fs.price),2) AS avg_product_price
FROM main_workspace.gold.fact_sales fs
JOIN main_workspace.gold.dim_products p
ON fs.product_id = p.product_id
JOIN main_workspace.silver.product_category_translation t
ON p.product_category_name = t.product_category_name
GROUP BY p.product_category_name, t.product_category_name_english
ORDER BY avg_product_price DESC
LIMIT 10
```
