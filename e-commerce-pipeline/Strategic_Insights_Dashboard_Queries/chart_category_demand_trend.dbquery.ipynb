```sql
SELECT
DATE_TRUNC('month', o.order_purchase_timestamp) AS order_month,

CONCAT(
INITCAP(REPLACE(p.product_category_name,'_',' ')),
'  ( ',
INITCAP(REPLACE(t.product_category_name_english,'_',' ')),
' )'
) AS product_category,

QUARTER(o.order_purchase_timestamp) AS order_quarter,

COUNT(fs.product_id) AS items_sold

FROM main_workspace.gold.fact_sales fs
JOIN main_workspace.gold.fact_orders o
ON fs.order_id = o.order_id
JOIN main_workspace.gold.dim_products p
ON fs.product_id = p.product_id
JOIN main_workspace.silver.product_category_translation t
ON p.product_category_name = t.product_category_name

GROUP BY
DATE_TRUNC('month', o.order_purchase_timestamp),
QUARTER(o.order_purchase_timestamp),
p.product_category_name,
t.product_category_name_english

ORDER BY order_month
```
