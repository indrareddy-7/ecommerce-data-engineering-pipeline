```sql
WITH customer_revenue AS (
SELECT
o.customer_id,
SUM(fs.price + fs.freight_value) AS revenue
FROM main_workspace.gold.fact_sales fs
JOIN main_workspace.gold.fact_orders o
ON fs.order_id = o.order_id
GROUP BY o.customer_id
),

top_customers AS (
SELECT
SUM(revenue) AS top_revenue
FROM (
SELECT revenue
FROM customer_revenue
ORDER BY revenue DESC
LIMIT 10
)
),

total_revenue AS (
SELECT SUM(revenue) AS total_rev
FROM customer_revenue
)

SELECT
ROUND((top_revenue / total_rev) * 100,2) AS top_customer_revenue_share
FROM top_customers, total_revenue
```
