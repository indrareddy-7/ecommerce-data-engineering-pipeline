```sql
SELECT
CASE
WHEN total_spend >= 500 THEN 'High Value Customers'
WHEN total_spend BETWEEN 200 AND 499 THEN 'Medium Value Customers'
ELSE 'Low Value Customers'
END AS customer_segment,
COUNT(*) AS customer_count
FROM (
SELECT
o.customer_id,
SUM(fs.price + fs.freight_value) AS total_spend
FROM main_workspace.gold.fact_sales fs
JOIN main_workspace.gold.fact_orders o
ON fs.order_id = o.order_id
GROUP BY o.customer_id
) t
GROUP BY customer_segment
```
