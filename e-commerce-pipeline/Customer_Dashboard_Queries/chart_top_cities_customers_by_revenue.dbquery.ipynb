```sql
SELECT
c.customer_city,
c.customer_state,
SUM(fs.price + fs.freight_value) AS total_revenue
FROM main_workspace.gold.fact_sales fs
JOIN main_workspace.gold.fact_orders o
ON fs.order_id = o.order_id
JOIN main_workspace.gold.dim_customers c
ON o.customer_id = c.customer_id
GROUP BY c.customer_city, c.customer_state
ORDER BY total_revenue DESC
LIMIT 10
```
